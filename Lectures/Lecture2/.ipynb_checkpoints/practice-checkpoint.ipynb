- `eval_expr(<(3 + 2) * 5>,{})`
    1. `<eval_expr(<25>,{})>`

SOLUTION OF EXERCISE 4
```
eval (<P>,S) -> <P'>, S'

{ x := 5, y := 1, z := 3}

z := x * 3 + y;
x := 3 * 5 - z;
y := z * 4 + 3 * x;
a := z - 3;
done;


eval(<z := x * 3 + y; x := 3 * 5 - z; y := z * 4 + 3 * x; a := z - 3; done>, { x := 5, y := 1, z := 3 })
  -> <z := 5 * 3 + y; x := 3 * 5 - z; y := z * 4 + 3 * x; a := z - 3; done>, { x := 5, y := 1, z := 3 }


eval(<z := 5 * 3 + y; x := 3 * 5 - z; y := z * 4 + 3 * x; a := z - 3; done>, { x := 5, y := 1, z := 3 })
  -> <z := 15 + y; x := 3 * 5 - z; y := z * 4 + 3 * x; a := z - 3; done>, { x := 5, y := 1, z := 3 }


eval(<z := 15 + y; x := 3 * 5 - z; y := z * 4 + 3 * x; a := z - 3; done>, { x := 5, y := 1, z := 3 })
  -> <z := 15 + 1; x := 3 * 5 - z; y := z * 4 + 3 * x; a := z - 3; done>, { x := 5, y := 1, z := 3 }


eval(<z := 15 + 1; x := 3 * 5 - z; y := z * 4 + 3 * x; a := z - 3; done>, { x := 5, y := 1, z := 3 })
  -> <z := 16; x := 3 * 5 - z; y := z * 4 + 3 * x; a := z - 3; done>, { x := 5, y := 1, z := 3 }


eval(<z := 16; x := 3 * 5 - z; y := z * 4 + 3 * x; a := z - 3; done>, { x := 5, y := 1, z := 3 })
  -> <done; x := 3 * 5 - z; y := z * 4 + 3 * x; a := z - 3; done>, { x := 5, y := 1, z := 16 }


eval(<done; x := 3 * 5 - z; y := z * 4 + 3 * x; a := z - 3; done>, { x := 5, y := 1, z := 16 }) 
  -> <x := 3 * 5 - z; y := z * 4 + 3 * x; a := z - 3; done>, { x := 5, y := 1, z := 16 }


eval(<x := 3 * 5 - z; y := z * 4 + 3 * x; a := z - 3; done>, { x := 5, y := 1, z := 16 })
  -> <x := 15 - z; y := z * 4 + 3 * x; a := z - 3; done>, { x := 5, y := 1, z := 16 }


eval(<x := 15 - z; y := z * 4 + 3 * x; a := z - 3; done>, { x := 5, y := 1, z := 16 })
  -> <x := 15 - 16; y := z * 4 + 3 * x; a := z - 3; done>, { x := 5, y := 1, z := 16 }


eval(<x := 15 - 16; y := z * 4 + 3 * x; a := z - 3; done>, { x := 5, y := 1, z := 16 })
  -> <x := -1; y := z * 4 + 3 * x; a := z - 3; done>, { x := 5, y := 1, z := 16 }


eval(<x := -1; y := z * 4 + 3 * x; a := z - 3; done>, { x := 5, y := 1, z := 16 })
  -> <done; y := z * 4 + 3 * x; a := z - 3; done>, { x := -1, y := 1, z := 16 }


eval(<done; y := z * 4 + 3 * x; a := z - 3; done>, { x := -1, y := 1, z := 16 })
  -> <y := z * 4 + 3 * x; a := z - 3; done>, { x := -1, y := 1, z := 16 }


eval(<y := z * 4 + 3 * x; a := z - 3; done>, { x := -1, y := 1, z := 16 })
  -> <y := 16 * 4 + 3 * x; a := z - 3; done>, { x := -1, y := 1, z := 16 }


eval(<y := 16 * 4 + 3 * x; a := z - 3; done>, { x := -1, y := 1, z := 16 })
  -> <y := 64 + 3 * x; a := z - 3; done>, { x := -1, y := 1, z := 16 }


eval(<y := 64 + 3 * x; a := z - 3; done>, { x := -1, y := 1, z := 16 })
  -> <y := 64 + 3 * -1; a := z - 3; done>, { x := -1, y := 1, z := 16 }


eval(<y := 64 + 3 * -1; a := z - 3; done>, { x := -1, y := 1, z := 16 })
  -> <y := 64 - 3 ; a := z - 3; done>, { x := -1, y := 1, z := 16 }


eval(<y := 64 - 3; a := z - 3; done>, { x := -1, y := 1, z := 16 })
  -> <y := 61; a := z - 3; done>, { x := -1, y := 1, z := 16 }


eval(<y := 61; a := z - 3; done>, { x := -1, y := 1, z := 16 })
  -> <done; a := z - 3; done>, { x := -1, y := 61, z := 16 }


eval(<done; a := z - 3; done>, { x := -1, y := 61, z := 16 })
  -> <a := z - 3; done>, { x := -1, y := 61, z := 16 }


eval(<a := z - 3; done>, { x := -1, y := 61, z := 16 })
  -> <a := 16 - 3; done>, { x := -1, y := 61, z := 16 }


eval(<a := 16 - 3; done>, { x := -1, y := 61, z := 16 })
  -> <a := 13; done>, { x := -1, y := 61, z := 16 }


eval(<a := 13; done>, { x := -1, y := 61, z := 16 })
  -> <done; done>, { x := -1, y := 61, z := 16, a:= 13 }


eval(<done; done>, { x := -1, y := 61, z := 16, a:= 13 })
  -> <done>, { x := -1, y := 61, z := 16, a:= 13 }
```
