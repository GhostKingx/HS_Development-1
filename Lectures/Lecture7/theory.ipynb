_Note_: ensure that students copy, by hand and on paper, the various programs written by the teacher on his machine. It is strongly advised to ask students *not* to use a laptop, as it will prove distracting.

- we will now perform some basic problem solving
- our goal is to implement some very primitive drawings by assembling strings
    - lines
    - lines with alternating symbols
    - squares
    - empty squares
    - triangles
    - pyramids
    - circles
- for each "problem", we will take the following steps:
    1. draw the figure
    2. analyse its elements; count the rows, identify the properties of each row (length, number of spaces, etc.)
    3. top-down, build a first raw program that prints the basic properties
    4. when we are satisfied that our program can identify the properties, then we move on to printing by accumulating on a string
- notice that we are in no hurry to start assembling the string (our actual output) yet
    - instead, we take our time exploring our problem domain
    - it is, in general, advisable to not work in a hurry to achieve one's result
    - very often, when performing engineering work, analysis and testing achieve results quicker than just trying the fastest route towards the goal
        - experts seem to skip the first steps, but this is not true: they have done it so often that they can do it all in their mind
- let us start with a single line:
    - the input is a number, `n`
    - the output is a line with a series of $n$ asterisks: `*...*`
    - a line with a series of $n$ asterisks is built by performing an instruction (adding an asterisk to a string), $n$ times
        - we found **the instructions that produce our output**
    - let us define a program that prints $n$ times: `I should add an asterisk`
    - let us verify it does, indeed, print $n$ times
    - let us now add the empty string and the string step
- let us move to a square:
    - the input is a number, `n`
    - the output is a square with a series of $n \times n$ asterisks: `*...*\n*...*\n`
    - each line is just a line of $n$ asterisks, like we just built a moment ago
        - so we actually have noticed that we do not need to solve the whole problem: we have to solve the problem of drawing a line $n$ times, but the actual line drawing we already know how to perform 
    - let us define a program that prints $n$ times: `I should add a line`
    - let us verify it does, indeed, print $n$ times
    - let us now add the empty string, and for the string step adapt (copy/paste) the code above
- let us now alter the square to an empty square (see figure below):
    - the empty square can be described as a full line, $n-2$ empty lines, and a full line again
    - let us build the skeleton that prints "I should add a line", "I should add an empty line", "I should add a line"
    - let us then add the code for the lines by adapting (copy/paste) the code from above
    - example figure for $n = 4$

```
****
*  *
*  *
****
```

- let us move to a triangle (see figure below):
    - to draw such a triangle, we must draw $n$ lines
        - the first line has length $1$
        - the second line has length $2$
        - ...
    - the skeleton program prints "I should add a line of length 1", "I should add a line of length 2", ...
    - let us then add the code for the lines
    - example figure for $n = 4$

```
*
**
***
****
```

- the last figure we draw in this lecture is the pyramid (see figure below):
    - this looks like a much more complex pattern
    - let us count, per line, what needs to be drawn
        - blank spaces, then stars
        - every line draws one space less, and two stars more
        - let us build the skeleton program that does this
    - how about the starting number of blanks ($b$) and stars ($s$)?
        - in the first line, the blanks are $n$, that is $b = n$
        - moreover, $s = 1$, since the pyramid always has a single star on top
    - let us now add the line drawing code
    - example figure for $n = 4$

```
   *
  ***
 *****
*******
```
